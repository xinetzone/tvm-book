# MXNet 模型可视化

```python
from mxnet.symbol import (
    Variable, FullyConnected,
    Activation, SoftmaxOutput
)
from mxnet.visualization import plot_network, print_summary

net = Variable('data')
net = FullyConnected(data=net, name='fc1', num_hidden=128)
net = Activation(data=net, name='relu1', act_type="relu")
net = FullyConnected(data=net, name='fc2', num_hidden=10)
net = SoftmaxOutput(data=net, name='out')
digraph = plot_network(net, shape={'data': (100, 200)},
                       node_attrs={"fixedsize": "false"})
digraph
```

```python
print_summary(net, shape={'data': (100, 200)})
```
