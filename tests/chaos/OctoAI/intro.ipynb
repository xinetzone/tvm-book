# OctoAI 简介

## Python SDK安装与配置

目前，SDK支持Python版本3.8.1及以上。强烈建议您使用如Conda或venv这样的虚拟环境来管理您的开发环境中的Python包。这有助于避免与当前项目无关或与系统依赖项冲突的已安装包之间的不兼容依赖性。

您可以在PyPi上查看Python SDK的完整发布历史以获取最新版本。

对于Mac和Linux，请使用 `pip` 安装octoai sdk Python包。强烈建议您在 `virtualenv` 中安装 OctoAI，以避免与系统包发生冲突。
```bash
python3 -m pip install octoai octoai-sdk -i https://pypi.tuna.tsinghua.edu.cn/simple 
```

## 将 API 令牌设置为环境变量
为了访问端点，请创建 [OctoAI API 令牌](https://octo.ai/docs/getting-started/how-to-create-octoai-access-token)。在设置环境变量的地方（例如您的 `.bashrc` 或 `.env` 文件）将 `OCTOAI_TOKEN` 设置为令牌值。

```bash
export OCTOAI_TOKEN=YOUR_TOKEN_HERE
```

当你实例化客户端时，它会自动检测 `OCTOAI_TOKEN` 作为环境变量并为你设置。

````python
from octoai.client import OctoAI

client = OctoAI()
````

或者，在创建 `OctoAI` 类时，你可以设置你的令牌；客户端也接受一个路径，该路径指向你在 `config_path` 变量中存储的API令牌。更多信息请参阅OctoAI文档。

```python
from octoai.client import OctoAI

client = OctoAI(api_key="YOUR_OCTOAI_API_TOKEN_HERE")
```
